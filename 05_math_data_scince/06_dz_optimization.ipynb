```python
def f(x, a, b):
    return np.e**a * np.sin(b*x) + x
```

```python
"""
method{‘lm’, ‘trf’, ‘dogbox’}, optional
Method to use for optimization. See least_squares for more details. 
Default is ‘lm’ for unconstrained problems and ‘trf’ if bounds are provided. 
The method ‘lm’ won’t work when the number of observations is less than the number of variables, use 
‘trf’ or ‘dogbox’ in this case.
"""

param_bounds = ([0,0],[5,5])
# param_bounds = ([-15,-15],[10,10])
res_lm = curve_fit(f,x,y,method='lm',p0=[2,2]) 
#popt, pcov = curve_fit(func, xdata, ydata)
# res[0] - п"олученные a,b
res_trf = curve_fit(f,x,y, method='trf', bounds=param_bounds) #,p0=[2,2]
# ,без ограничения расходится:
# RuntimeError: Optimal parameters not found: The maximum number of function evaluations is exceeded.
res_dogbox = curve_fit(f,x,y, method='dogbox', bounds=param_bounds) #p0=[2,2]
# RuntimeError: Optimal parameters not found: The maximum number of function evaluations is exceeded.
print(res_trf)
print()
print(res_dogbox)
print()
print(res_lm)
```
