```python
# f(x) = x**5 + 4*sin(2*x) + cos(3*x+3)
# производная:

# 5 * x**4 + 4*cos(2*x)*2 - sin(3*x + 3) * 3
# 5*x**4 + 8 * cos(2*x) -3sin(3x+3)
# вторая производная:

# 20 * x**3 + 8* (-sin(2*x)) * 2 -3*cos(3x + 3)*3
# 20 * x **3 - 16 sin(2x) - 9*cos(3*x + 3)

# прверю себя:
x = symbols('x')
di_1 = diff(x**5 + 4*sin(2*x) + cos(3*x+3))
di_2 = diff(5*x**4 - 3*sin(3*x + 3) + 8*cos(2*x))
print('Первая производная:')
print(di_1)
print('Вторая производная:')
print(di_2)
print('значение производных в точке x = 1')
print('Первая')
print(di_1.subs(x, 1))
difff_1 = diff(x**5 + 4*sin(2*x) + cos(3*x + 3)).evalf(subs={'x':1})
print(difff_1)
print('Вторая')
print(di_2.subs(x, 1))
diff_2 = diff(diff(x**5 + 4*sin(2*x) + cos(3*x + 3))).evalf(subs={'x':1})
print(diff_2)
```
