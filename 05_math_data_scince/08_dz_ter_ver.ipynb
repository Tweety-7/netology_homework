Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

```python
"""
если рассматривать все три события - они представляют собой уже условную вероятность
В случае наступления двух событий третье автоматически становится верным, и наоборот
т.е. при наступление очередного удачного события, вероятность всех трёх будет увеличиваться.
"""
# Вероятность наступления каждого из событий А:
# P(A_1) = P(A_2) = P(A_3) =
# 1/2 * 1/2 = 1/4
# 1/4 = 0.25
# для выполнения условия независимости в совокупности должно выполняться условие: P(A_1) =  P(A_2) = P(A_3) = P(A_1A_2A_3)
# а оно не выполняется. тк P(A_1A_2A_3) = P(A_1)*P(A_2)*P(A_3) = 1/4 * 1/4 * 1/4
# 1/64 = > не являются независимыми в совокупности
```

```python
# 1. нормальное распределение
# loc - среднее
# scale - стандартное отклонение
size_0 = 1000
norm_1 = stats.norm(loc=1, scale=1,).rvs(size = size_0)
norm_2 = stats.norm(loc=2, scale=1).rvs(size = size_0)
norm_3 = stats.norm.rvs(loc=1, scale=7, size = size_0)
norm_4 = stats.norm.rvs(loc=-2, scale=7, size = size_0)
norm_5 = stats.norm.rvs(loc=-2, scale=1, size = size_0)
```

```python
plt.hist(norm_1)
plt.hist(norm_2,alpha=0.5,color='red');
# по сравнению с norm_1 съехало вправо. т.к. среднее изменилось с 1 до 2
```

```python
plt.hist(norm_1)
plt.hist(norm_3,alpha=0.5,color='red');
# по сравнению с norm_1 растянулось. т.к. отклонение выросло с 1 до 7
```

```python
plt.hist(norm_3)
plt.hist(norm_4,alpha=0.5,color='red');
# центр сместился влево (с 1 до -2)
```

```python
plt.hist(norm_4)
plt.hist(norm_5,alpha=0.5,color='red');
# уменьшила разброс с 7 до 1
```

```python
# функцию плотности распредления
pdf_norm_1 = stats.norm.pdf(norm_1)
pdf_norm_2 = stats.norm.pdf(norm_2)
pdf_norm_3 = stats.norm.pdf(norm_3)
pdf_norm_4 = stats.norm.pdf(norm_4)
pdf_norm_5 = stats.norm.pdf(norm_5)
plt.plot(norm_1,pdf_norm_1,'.',markersize=15)
plt.plot(norm_2,pdf_norm_2,'r.')
# начало съехало вправо из-за увеличения среднего
```

```python
plt.plot(norm_1,pdf_norm_1,'.',markersize=20)
plt.plot(norm_3,pdf_norm_3,'r.')
# растянулась вдоль оси, из-за увеличения отклонения
```

```python
plt.plot(norm_3,pdf_norm_3,'.',markersize=20)
plt.plot(norm_4,pdf_norm_4,'r.')
# по графику видно - съехала левее
```

```python
plt.plot(norm_3,pdf_norm_3,'.',markersize=20)
plt.plot(norm_5,pdf_norm_5,'r.')
# уменьшился разброс.хоть кто-то похудел к лету =)
```
