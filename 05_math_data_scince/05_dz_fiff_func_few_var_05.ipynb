Задание 3

Постройте трехмерный график функции

```python
# f(x) = x_2**2 - (x_1**2)* x_2 + 2*x_1 * x_2
```

```python
# трёхмерный график
def fun_gr(x_1,x_2):
  return x_2**2 - (x_1**2) * x_2 + 2*x_1*x_2
fig = plt.figure()
ax = fig.gca(projection='3d')
x_1 = np.arange(-5,5,0.25)
x_2 = np.arange(-5,5,0.25)
X_1, X_2 = np.meshgrid(x_1,x_2)
F = fun_gr(X_1,X_2)

surf = ax.plot_surface(X_1,X_2,F,cmap = cm.coolwarm,linewidth = 0, antialiased = False)
ax.set_zlim(-150,200.0)
ax.zaxis.set_major_formatter(FormatStrFormatter('%.0f'))
fig.colorbar(surf, shrink = 0.5, aspect=5)
ax.azim = 10
ax.dist = 10
ax.elev = 50
plt.show()
```
