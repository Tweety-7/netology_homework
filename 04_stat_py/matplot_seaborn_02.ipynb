```python
df_t = pd.read_csv('totals_stats.csv')
df_t.head(23)
# df_t.info()
```

```python
# 1. посмотрим кто был результативнее по очкам
result_by_player = df_t[['2P','3P','FT','MP','AST','Player']].groupby('Player').sum().reset_index()
# df_t.head()
result_by_player
```

```python
plt.plot(result_by_player['Player'], result_by_player[['2P', '3P', 'FT','AST']])
plt.xlabel('Игроки')
plt.ylabel('кол-во забитых')
plt.legend(['2P', '3P', 'FT', 'AST']);
```

```python
# из графика видно, что MJ был результативнее других в двухочковых 
# и попаданиях со штрафной
# однако молот Леброн остаётся лучшим снайпером из-за дуги(3-очка)
# и лучшим напарником для голевой передачи

# из графика понятно что все звёзды 2-х очковых бросков реализовали наибольшее количество
```

```python
# распределение времени игры для каждого игрока
KB_min = df_t['MP'][df_t['Player'] == 'Kobe Bryant']
LJ_min = df_t['MP'][df_t['Player'] == 'Lebron James']
MJ_min = df_t['MP'][df_t['Player'] == 'Michael Jordan']

KB_min.plot(kind = 'hist', bins = 7, alpha = 0.7, density=True)
LJ_min.plot(kind = 'hist', bins = 7, alpha = 0.5, density=True)
MJ_min.plot(kind = 'hist', bins = 7, alpha = 0.5, density=True)
plt.legend(['KB', 'LJ','MJ'])
# print(len(KB_min), len(LJ_min), len(MJ_min))
```

```python
# не очень конечно достоверно, но выглядит так будто Джордан в большем кол-ве случаев дольше и меньше находился.
# раньше можно было увидеть, что суммарного времени игры у него незначительно меньше, чем у остальных
```

```python
# потери
losses_by_players = df_t[['TOV', 'PF','Player']]
```

```python
MJ_KB_losses.groupby('Player').sum().plot(kind = 'bar')
# по графику можно предположить, что отрицательных действий Коби совершил больше
# так же видно что кол-во потерь привосходит кол-ву персональных фолов и у одного и у второго игрока
```
