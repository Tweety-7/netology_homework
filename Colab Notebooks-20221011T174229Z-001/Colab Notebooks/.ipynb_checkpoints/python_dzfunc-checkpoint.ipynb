```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
def show_p():
  number_doc = str(input('Введите номер документа: '))
  for doc in documents:
    if str(doc['number']) == str(number_doc):
      print(f"Владелец документа: {doc['name']}")
      return
  print("Документ не найден в базе")
  return

def shelf_s(num_doc):
  # print(f" kek {num_doc}")
  if num_doc == 'nothing': 
    num_doc = input('Введите номер документа: ')
  for shel in directories.keys():
    if str(num_doc) in directories[shel]:
      # print(f"Документ хранится на полке: {shel}")
      return shel
  # print('обошла for')
  return 0

def show_all_l():
  for one in documents:
    print(f"№: {one['number']}, тип: {one['type']}, владелец: {one['name']}, полка хранения {shelf_s(num_doc = one['number'])}")

def add_shel_ads():
   new_num_shel = input("Введите номер полки:")
   if new_num_shel in directories.keys():
     print(f"Такая полка уже существует. Текущий перечень полок: {', '.join(directories)}")
   else:
    directories[str(new_num_shel)] = []
    print(f"Полка добавлена. Текущий перечень полок: {', '.join(directories)}")

def del_shel_ds():
  num_shel = input('Введите номер полки:')
  if num_shel not in directories.keys():
    print(f"Такой полки не существует. Текущий перечень полок: {', '.join(directories)}.")
  elif directories[num_shel] == []:
    del directories[num_shel]
    print(f"Полка удалена. Текущий перечень полок:{', '.join(directories)}.")
  else:
    print(f"На полке есть документа, удалите их перед удалением полки. Текущий перечень полок:{', '.join(directories)}")
  return

def add_doc_ad():
  num_doc = input('Введите номер документа:')
  type_doc = input('Введите тип документа:')
  name = input('Введите владельца документа:')
  num_shel = input('Введите полку для хранения:')
  if num_shel in directories.keys():
    directories[num_shel].append(num_doc)
    one_doc = {'type': type_doc, 'number': num_doc, 'name': name}
    documents.append(one_doc)
    print('Документ добавлен.', end = ' ')
  else:
    print('Такой полки не существует. Добавьте полку командой ads.')

def del_doc_d():
  num_doc = input('Введите номер документа:')
  for one_doc in documents:
    if one_doc['number'] == num_doc:
      documents.remove(one_doc)
      print('Документ удален.')
      return
  print('Документ не найден в базе.')
  return    
def move_m():
  num_doc = input('Введите номер документа:')
  old_shel = shelf_s(num_doc)
  if old_shel:
    new_shel = input('Введите номер полки:')
    if new_shel not in directories.keys():
      print(f"Такой полки не существует. Текущий перечень полок: {', '.join(directories)}.")
      return
    directories[new_shel].append(num_doc)
    # old_shel_idx = directories[old_shel].index(num_doc)
    directories[old_shel].remove(num_doc)

  else:
    print('Документ не найден в базе')
    return
def main(command):
  while command != 'q':
    command = input("Введите команду:")
    if command == 'p':
      show_p()
    elif command == 's':
      # shel = shelf_s(10006)
      shel = shelf_s('nothing')
      # print(f"voot shel: {shel}")
      if shel:
        print(f"Документ хранится на полке: {shel}")
      else:
        print("Документ не найден в базе")
    elif command == 'l':
      show_all_l()
    elif command == 'ads':
      add_shel_ads()
    elif command == 'ds':
      del_shel_ds()
    elif command == 'ad' or command == 'd' or command == 'm':
      if command == 'ad':
        add_doc_ad()
      elif command == 'd':   
        del_doc_d()
      else: #command = m
        move_m()
      print('Текущий список документов:')
      show_all_l()

      


  return  
main( command = 'z')


```
