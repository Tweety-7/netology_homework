```python
import re

"""
Задание 1
Напишите функцию, 
которая принимает на вход строку и проверяет 
является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). 
Обратите внимание, что не все буквы кириллического алфавита 
используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.
"""

def valid_number(num_str):
  valid_id = '([ABEKMHOPCTYX]{1}\d{3}[ABEKMHOPCTYX]{2})(\d{2,3})'
  valid = re.findall(valid_id,num_str)

  if valid:
    print(f"Номер {valid[0][0]} валиден. регион: {valid[0][1]}")
  else:
    print('Номер не валиден')
# engl ABEKMHOPCTYX
# rus АВЕКМНОРСТУХ
car_id = 'A222BC96'
valid_number(car_id)
car_id = 'АБ22ВВ193'
valid_number(car_id)


```

```python
"""
Задание 2
Напишите функцию, 
которая будет удалять
 все последовательные 
 повторы слов из заданной строки при помощи регулярных выражений.
"""
def del_repeat(str_start):
  rep_word = r'([а-яёА-ЯЁ]+)\s(\1)+'
  while re.findall(rep_word, str_start):
    str_start = re.sub(rep_word, r'\1', str_start)
  print(str_start)


some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений.'
del_repeat(some_string)
```

```python
"""
Задание 3
Напишите функцию, 
которая будет возвращать 
акроним по переданной в нее строке со словами.
"""
def acronim(str_start):
  word = r'[а-яёА-ЯЁa-zA-Z]+'
  str_fin = ''
  list_str = re.findall(word,str_start)
  for i in list_str:
    str_fin += i[0].upper()
  print(str_fin)

some_words = 'Информационные технологии'
acronim(some_words)
some_words = 'Near Field Communication'
acronim(some_words)
```

```python
"""
Задание 4
Напишите функцию, которая будет принимать на вход список 
email-адресов и выводить их распределение по доменным зонам.
"""
def domen_zone(list_adress):
  mail_pattern = r'\w+@(\w+|).(\w{2,3})'
  domen_dict = {}
  for adress in list_adress:
    domen = re.findall(mail_pattern, adress)
    if domen is not None:
      # если домен существует добавим в словарь
      domen_name = domen[0][0] + '.' + domen[0][1]
      domen_dict[domen_name] = domen_dict.get(domen_name, 0) + 1
  # печать всех доменов
  for key, value in domen_dict.items():
    print(f'{key}: {value}') 


emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
domen_zone(emails)
```

```python
"""
Задание 5 (необязательное)
Напишите функцию, 
которая будет подсчитывать 
сколько слов начинается на гласные, 
а сколько на согласные буквы в тексте 
(текст может быть написан как с использованием букв кириллицы, так и латиницы).
"""
def consonants_vowels(text):
  consonants_vowels_dict = {}
  # # согласные - все остальные, что не совсем верно для цифр и прочего
  # parretn_consonants = r'[]'
  # гласные
  pattern_vowels = r'^[аоуэыяёюеиАОУЭЫЯЁЮИaeiuyoAEIUYO]{1}'
  pattern_all = r'[а-яёА-ЯЁa-zA-Z]'
  # word = r'[а-яёА-ЯЁa-zA-Z]'
  text_split = re.split(r'\s', text)
  for word in text_split:
    find_all = re.findall(pattern_all,word)
    if find_all is not None:
      find_vowels = re.findall(pattern_vowels, word)
      if find_vowels :
        # нашли гласную
        consonants_vowels_dict['vowels'] = consonants_vowels_dict.get('vowels',0) + 1
      else:
        consonants_vowels_dict['consonants'] = consonants_vowels_dict.get('consonants',0) + 1
  for i in consonants_vowels_dict.keys():
    if i == 'vowels':
      print(f"Слов на гласные буквы: {consonants_vowels_dict['vowels']}")
    else:
      print(f"Слов на согласные буквы: {consonants_vowels_dict['consonants']}")





some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'
consonants_vowels(some_text)
```
