```python
from datetime import datetime
str_the_moscow_times = "Wednesday, October 2, 2002"

str_the_guardian = "Friday, 11.10.13"
str_daily_news = "18 August 1977"
'''
Для каждой газеты из списка напишите формат указанной даты для перевода в объект datetime:
'''
date_datetime_moscow = datetime.strptime(str_the_moscow_times, "%A, %B %d, %Y")
date_datetime_guard = datetime.strptime(str_the_guardian, "%A, %d.%m.%y")
date_datetime_daily = datetime.strptime(str_daily_news, "%d %B %Y")
print(date_datetime_moscow)
print(date_datetime_guard)
print(date_datetime_daily)
```

```python
from datetime import timedelta

def date_range(start_date, end_date):
  d = []
  try:
    start_date_datetime = datetime.strptime(start_date, "%Y-%m-%d")
    end_date_datetime = datetime.strptime(end_date, "%Y-%m-%d")
    current_date = start_date_datetime
    while current_date <= end_date_datetime:
      d.append(current_date)
      current_date += timedelta(days = 1)
  except:
    pass
  finally:
    return d

print(date_range("2022-08-27", "2022-08-31"))
```
