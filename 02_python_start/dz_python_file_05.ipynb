```python
'''
Переведите содержимое файла purchase_log.txt в словарь purchases вида:
{‘1840e0b9d4’: ‘Продукты’, ...}
'''

with open('purchase_log.txt', 'r') as f:
  file_purchase = f.readlines()
import json 
dict_purchase = {}
# with open('purchase_log.txt', 'r') as f:
for line in file_purchase:
    line_d = json.loads(line)
    dict_purchase[line_d['user_id'].strip()] = line_d['category']
keys_purchace = dict_purchase.keys()
```

```python
'''
Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файлfunnel.csv визиты из файла visit_log.csv, 
в которых были покупки с указанием категории.
'''

with open('visit_log.csv', 'r') as f:
  for line in f:
    id = line.strip().split(',')[0]
    id = id.strip()
    if id in keys_purchace:
      line_write = line.strip() + ',' + dict_purchase[id] + '\n'
      with open('funnel.csv', 'a') as f_a:
        f_a.writelines(line_write)

    
```
