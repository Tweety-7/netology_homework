```python
# Задание 2
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт'
]
dict_quer = {}
for i in queries:
  len_i = len(i.split(' '))
  dict_quer[len_i] = dict_quer.get(len_i,0) + 1
# print(dict_quer)
key_list = list(dict_quer.keys())
key_list.sort()
for k in key_list:
  print(f"Поисковых запросов, содержащих {k} слов(а): {round(dict_quer[k] *100/len(queries),2)}%")

```

```python
# Задание 3
results = {
    'vk': {'revenue': 103, 'cost': 98},
    'yandex': {'revenue': 179, 'cost': 153},
    'facebook': {'revenue': 103, 'cost': 110},
    'adwords': {'revenue': 35, 'cost': 34},
    'twitter': {'revenue': 11, 'cost': 24},
}
for name, past_res in results.items():
  # print(name)
  # print(past_res)
  roi = round((past_res['revenue']/past_res['cost'] - 1)*100, 2)
  past_res['ROI'] = roi
  results[name] = past_res
  print(f"{name}: {results[name]}")
```

```python
# Задание 6
cook_book = {
  'салат': [
     {'ingridient_name': 'сыр', 'quantity': 50, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 2, 'measure': 'шт'},
     {'ingridient_name': 'огурцы', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'маслины', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'оливковое масло', 'quantity': 20, 'measure': 'мл'},
     {'ingridient_name': 'салат', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'перец', 'quantity': 20, 'measure': 'гр'}
    ],
  'пицца': [
     {'ingridient_name': 'сыр', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'колбаса', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'бекон', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'оливки', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'тесто', 'quantity': 100, 'measure': 'гр'},   
    ],
  'лимонад': [
     {'ingridient_name': 'лимон', 'quantity': 1, 'measure': 'шт'},
     {'ingridient_name': 'вода', 'quantity': 200, 'measure': 'мл'},
     {'ingridient_name': 'сахар', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'лайм', 'quantity': 20, 'measure': 'гр'},    
    ]
}
porc = int(input("Введите количество порций: "))
# porc = 3
ingridients = {}
for dish in cook_book.keys():
  one_dish = cook_book[dish] #список словарей ингридиентов
  for one in one_dish: #по каждому словарю ингридиента
    # print(one['ingridient_name'])
    name = one['ingridient_name']
    measure = one['measure']
    quantity = int(one['quantity']) * porc
    one_ingridients = ingridients.get(name,{})
    one_ingridients[measure] = one_ingridients.get(measure, 0) + quantity
    ingridients[name] = one_ingridients
for key in ingridients.keys():
  mini_dict = ingridients[key]
  for k in mini_dict.keys():
    print(f"{key}: {mini_dict[k]} {k}")


# !!!!!!!!!!! ох чую как-то легче всё делается!
# но зато у меня удобнее собрать карзину, помидоры следуют друг за другом
# не придётся дважды возвращаться за ними =Р

# print(ingridients)
```
